```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python
data={
    'Cheap':['yes','yes','no','yes','no'],
    'Promo':['yes','no','yes','yes','no'],
    'Buy?':['yes','yes','no','yes','no']
}
```

```python
df=pd.DataFrame(data)
```

```python
#step2:encode categorical variable
le_cheap=LabelEncoder()
le_promo=LabelEncoder()
le_buy=LabelEncoder()
```

```python
df['Cheap']=le_cheap.fit_transform(df['Cheap']) #yes=1,no=0
df['Promo']=le_promo.fit_transform(df['Promo'])
df['Buy?']=le_buy.fit_transform(df['Buy?'])
```

```python
#step 3: feature (X) and target(y)
X=df[['Cheap','Promo']]
y=df['Buy?']
```

```python
#step 4: train naive bayes
model=CategoricalNB()
model.fit(X,y)
```

```python
#step 5: predict on the same data
predictions=model.predict(X)
```

```python
#step 6:show result with decode labels
df['Predicted_Buy']=le_buy.inverse_transform(predictions)
df['Actual_Buy']=le_buy.inverse_transform(y)
```

```python
print(df)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y,predictions)
print(report)
```

# Dataset-2

```python
data1={
    'Weather':['sunny','sunny','overcast','rain','rain'],
    'Wind':['weak','strong','weak','weak','strong'],
    'Play?':['yes','no','yes','yes','no']
}
```

```python
df=pd.DataFrame(data1)
```

```python
le_weather=LabelEncoder()
le_wind=LabelEncoder()
le_play=LabelEncoder()
```

```python
df['Weather']=le_weather.fit_transform(df['Weather']) #yes=1,no=0
df['Wind']=le_wind.fit_transform(df['Wind'])
df['Play?']=le_play.fit_transform(df['Play?'])
```

```python
X=df[['Weather','Wind']]
y=df['Play?']
```

```python
model=CategoricalNB()
model.fit(X,y)
```

```python
predictions=model.predict(X)
```

```python
df['Predicted_play']=le_play.inverse_transform(predictions)
df['Actual_play']=le_play.inverse_transform(y)
```

```python
print(df)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y,predictions)
print(report)
```
